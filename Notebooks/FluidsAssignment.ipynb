```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

## Streakline

```python
f = lambda y,t: np.exp(np.log(y)+(2*t*np.log(y)) - np.log(y)**2)
plt.figure(figsize=(10,6))
y = np.linspace(0.00001,4,200)
time = np.linspace(0,2,5)
linestyles = ['solid', 'dashed', 'dashdot', 'dotted', (0, (3, 1, 1, 1))]
for i in range(len(time)):
    x = f(y,time[i])
    plt.plot(x,y,label = f't = {time[i]}',ls =linestyles[i],lw=2.5)
plt.grid()
plt.xlim(-1,15)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Streakline')
plt.scatter(x = 1, y = 1, color = 'r', zorder = 1 , label ='(x,y) = (1,1)')
plt.legend()
plt.show()
```

## StreamLine

```python
f = lambda y,t : np.pow(y,2*t+1)
plt.figure(figsize=(10,6))
y = np.linspace(0,1.5,2000)
time = np.linspace(0,2,5)
for i in range(len(time)):
    x = f(y,time[i])
    plt.plot(x,y,label = f't = {time[i]}',ls =linestyles[i],lw = 2.5)
plt.scatter(x = 1, y = 1, color = 'r', zorder = 1 , label ='(x,y) = (1,1)')
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Streamlines')
plt.legend()
plt.show()
```

## PathLines

```python
t = np.linspace(0,2,2000)
fx = lambda t:np.exp(t+t**2)
fy = lambda t:np.exp(t)
x = fx(t)
y = fy(t)
plt.figure(figsize=(10,6))
plt.scatter(x = 1, y = 1, color = 'r', zorder = 1 , label ='(x,y) = (1,1)',lw = 2.5)
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(0,75)
plt.ylim(0,7)
plt.title('Pathline')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,8))
f = lambda y,t: np.exp(np.log(y)+(2*t*np.log(y)) - np.log(y)**2)

y = np.linspace(1,4,200)
time = np.linspace(0,2,5)
for i in range(len(time)):
    x = f(y,time[i])
    plt.plot(x,y,label = f't = {time[i]} streakline',ls ='--')

f = lambda y,t : np.pow(y,2*t+1)
y = np.linspace(0,4,2000)
time = np.linspace(0,2,5)
for i in range(len(time)):
    x = f(y,time[i])
    plt.plot(x,y,label = f't = {time[i]} streamline',ls ='-')
plt.scatter(x = 1, y = 1, color = 'r', zorder = 1 , label ='(x,y) = (1,1)')


t = np.linspace(0,2,2000)
fx = lambda t:np.exp(t+t**2)
fy = lambda t:np.exp(t)
x = fx(t)
y = fy(t)
plt.plot(x,y,label = 'pathline',ls = ":")

plt.ylim(0,4)
plt.xlim(-2,20)
plt.legend()
plt.grid()
plt.legend()
plt.show()
```
