```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

**Laplace equation in 2D**

$u_{xx} + u_{yy} = 0$

1. Length along x and y
2. grid points along x and y
3. dx and dy
4. Boundary conditions
5. Iterative formulation

```python
Lx =1.0
Ly = 1.0

nx =50
ny =50

x = np.linspace(0,Lx,nx+2)
print(x)

y = np.linspace(0,Ly,ny+2)
print(y)


dx = Lx/(nx+1)
dy = Ly/(ny+1)

#Temperature at the boundary
Tleft = 0.
Tbottom = 0.
Tright = 100.
Ttop =100.


#setup the boundary conditions

T = np.zeros((nx+2,ny+2))
print(T)

Tguess = 10.
T.fill(Tguess)
T
#Boundary conditions
T[0,:] = Tleft
T[-1,:] = Tright
T[:,-1] = Ttop
T[:,0] = Tbottom
T

print(T)
maxiter =1000

for _ in range(maxiter):
    for i in range(1,nx+1):
        for j in range(1,ny+1):
            T[i,j] = ((dy**2)*(T[i+1,j]+T[i-1,j]) + (dx**2)*(T[i,j+1]+T[i,j-1]))/(2*(dx**2+ dy**2))
print(T)
```

```python
X,Y = np.meshgrid(x,y,indexing='ij')
X
```

```python
plt.contourf(X,Y,T,levels=20,cmap='plasma')
plt.colorbar(label = 'Temperature')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(X,Y,T,cmap = 'hot')
ax.set_xlabel('X coordinates')
ax.set_ylabel('Y coordinates')
ax.set_zlabel('Temperature')
plt.show()
```
