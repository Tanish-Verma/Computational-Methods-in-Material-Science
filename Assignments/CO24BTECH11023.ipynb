## Area of circle

- Make a grid of div X div dimensions around the circle
- Iterate over x and y coordinate divisions
- Make a count var to count the number of boxes inside the circle
- Find the midpoint of each grid section
- calculate the distance of the midpoint from the center 
- if the distance is less than or equal to radius increment count
- finally multiply the count with the area of each box


```python
r =1
div = 10000
x = np.linspace(-1*r,r,div+1)
y = np.linspace(-1*r,r,div+1)
dx = x[1] - x[0]
dy = y[1] - y[0]
```

```python

count  =0
eq = lambda x,y : np.sqrt(x**2+y**2)
time_start = time.time()
for i in tqdm(range(div)):
    for j in range(div):
        xmid = (x[i]+x[i+1])/2
        ymid = (y[j]+y[j+1])/2
        if(eq(xmid,ymid)<=r):
            count = count+1
area = dx*dy*count
time_end = time.time()
print(f'Aprroximate area of circle = {area} , time taken = {time_end - time_start}')
```

```python
actualA = np.pi*r**2
print(f'relative error {np.abs(actualA - area)}')
```
